```python
from tensorflow.keras import layers, models, optimizers
from tensorflow.keras.callbacks import LearningRateScheduler
import numpy as np

def create_model(input_shape):
    model = models.Sequential([
        layers.Conv2D(32, (3, 3), activation='relu', input_shape=input_shape),
        layers.BatchNormalization(),
        layers.MaxPooling2D((2, 2)),
        layers.Dropout(0.2),
        
        layers.Conv2D(64, (3, 3), activation='relu'),
        layers.BatchNormalization(),
        layers.MaxPooling2D((2, 2)),
        layers.Dropout(0.3),
        
        layers.Conv2D(128, (3, 3), activation='relu'),
        layers.BatchNormalization(),
        layers.MaxPooling2D((2, 2)),
        layers.Dropout(0.4),
        
        layers.GlobalAveragePooling2D(),
        layers.Dense(256, activation='relu'),
        layers.Dropout(0.5),
        layers.Dense(28, activation='softmax')
    ])
    
    model.compile(
        optimizer='adam',
        loss='sparse_categorical_crossentropy',
        metrics=['accuracy']
    )
    return model

# Example usage:
input_shape = (50, 50, 3)  # Adjust input shape as needed
model = create_model(input_shape)

# Training parameters:
batch_size = 32
epochs = 50

# Learning rate scheduler
def lr_schedule(epoch, lr):
    if epoch > 30:
        lr = lr * 0.1
    return lr

callbacks = [LearningRateScheduler(lr_schedule)]

# Train the model
history = model.fit(X_train_scal, Y_train, batch_size=batch_size, epochs=epochs, validation_data=(X_test_scal, Y_test), callbacks=callbacks)

# Evaluate the model
loss, accuracy = model.evaluate(X_test_scal, Y_test)
print(f"Test Loss: {loss:.4f}, Test Accuracy: {accuracy:.4f}")
```

```python
import tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.optimizers import Adam
import numpy as np

def create_mobilenet_model(input_shape=(50, 50, 3), num_classes=28):
    # Load the pre-trained MobileNetV2 model without the top layer
    base_model = MobileNetV2(weights='imagenet', include_top=False, input_shape=input_shape)
    
    # Freeze the base model's layers to preserve the pre-trained weights
    base_model.trainable = False
    
    # Add your own classification layers
    model4 = models.Sequential([
        base_model,
        layers.GlobalAveragePooling2D(),
        layers.Dense(256, activation='relu'),
        layers.Dropout(0.5),
        layers.Dense(num_classes, activation='softmax')  # Adjust the number of classes as needed
    ])
    
    # Compile the model with an optimizer and loss function suitable for classification
    model4.compile(optimizer=Adam(), loss='sparse_categorical_crossentropy', metrics=['accuracy'])
    
    return model4

# Example usage:
# Assuming you have your training and test data in X_train, Y_train, X_test, and Y_test variables
input_shape = (50, 50, 3)  # Adjust input shape as needed
num_classes = 28  # Adjust the number of classes as needed

# Create the MobileNet model
model41 = create_mobilenet_model(input_shape=input_shape, num_classes=num_classes)

# Train the model
batch_size = 32
epochs = 50

```

```python
def create_resnet50_model(input_shape=(50, 50, 3), num_classes=28):
    # Load the pre-trained ResNet-50 model without the top layer
    base_model = ResNet50(weights='imagenet', include_top=False, input_shape=input_shape)
    
    # Freeze the base model's layers to preserve the pre-trained weights
    base_model.trainable = False
    
    # Add your own classification layers
    model3 = models.Sequential([
        base_model,
        layers.GlobalAveragePooling2D(),
        layers.Dense(256, activation='relu'),
        layers.Dropout(0.5),
        layers.Dense(num_classes, activation='softmax')  # Adjust the number of classes as needed
    ])
    
    # Compile the model with an optimizer and loss function suitable for classification
    model3.compile(optimizer=Adam(), loss='sparse_categorical_crossentropy', metrics=['accuracy'])
    
    return model3

# Example usage:
# Assuming you have your training and test data in X_train, Y_train, X_test, and Y_test variables
input_shape = (50, 50, 3)  # Adjust input shape as needed
num_classes = 28  # Adjust the number of classes as needed

# Create the ResNet-50 model
model3 = create_resnet50_model(input_shape=input_shape, num_classes=num_classes)

# Train the model
batch_size = 32
epochs = 50

# Train the model with your data (adjust X_train, Y_train, etc.)
history = model3.fit(X_train, Y_train, batch_size=batch_size, epochs=epochs, validation_data=(X_test, Y_test))

# Evaluate the model (adjust X_test, Y_test)
loss, accuracy = model3.evaluate(X_test, Y_test)
print(f"Test Loss: {loss:.4f}, Test Accuracy: {accuracy:.4f}")
```

```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.models import load_model

# Load the saved model
model_path = "D:/Projects/Tamil Character Recognition/new/code/cnn_model.h5"
model = load_model(model_path)

# Check the model's input shape
input_shape = model.input_shape

# Function to preprocess the image
def preprocess_image(image_path):
    # Load the image
    img = cv2.imread(image_path)
    if img is None:
        print(f"Error: Unable to load image from path: {image_path}")
        return None
    
    # Convert the image to the correct number of channels if needed
    if input_shape[-1] == 3:
        if len(img.shape) == 2:
            # Convert grayscale to RGB
            img = cv2.cvtColor(img, cv2.COLOR_GRAY2RGB)
    elif input_shape[-1] == 1:
        if len(img.shape) == 3:
            # Convert RGB to grayscale
            img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
            # Add an extra dimension to match the model's expected shape
            img = np.expand_dims(img, axis=-1)

    # Resize the image to the input size expected by the model
    img_resized = cv2.resize(img, (input_shape[2], input_shape[1]))

    # Normalize the image
    img_normalized = img_resized / 255.0

    # Expand the dimensions to match the input shape (batch size of 1)
    img_expanded = np.expand_dims(img_normalized, axis=0)

    return img_expanded


# Classify a new image
def classify_new_image(image_path):
    # Preprocess the image
    img = preprocess_image(image_path)
    if img is None:
        return

    # Make a prediction using the model
    prediction = model.predict(img)

    # Extract the predicted label
    predicted_label = np.argmax(prediction)

    return predicted_label

# Path to the new image you want to classify
new_image_path = "D:/Projects/Tamil Character Recognition/new/augumented_images/8/_0_1364.jpg"

# Classify the new image
predicted_label = classify_new_image(new_image_path)

# Print the predicted label
print(f"Predicted label for the new image: {predicted_label}")

# Optionally, you can visualize the new image
# Load the image in color for visualization
img_color = cv2.imread(new_image_path)
img_color_rgb = cv2.cvtColor(img_color, cv2.COLOR_BGR2RGB)  # Convert from BGR to RGB

# Display the image with the predicted label
plt.imshow(img_color_rgb)
plt.title(f"Predicted label: {predicted_label}")
plt.axis('off')
plt.show()


```
